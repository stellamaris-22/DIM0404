```python
import numpy as np
import sympy
import matplotlib.pyplot as plt
```

```python
x = np.arange(-50, 51)
y = np.cbrt(x) + x**3 + 7 
plt.title("a)") 
plt.xlabel("x") 
plt.ylabel("f(x)") 
plt.plot(x,y) 
plt.show()
```

```python
x = np.arange(-50, 51)
y = np.abs(x)
plt.title("b)") 
plt.xlabel("x") 
plt.ylabel("f(x)") 
plt.plot(x,y) 
plt.show()
```

```python
x = np.arange(-50, 51)
y = np.sin(x)
plt.title("c)") 
plt.xlabel("x") 
plt.ylabel("f(x)") 
plt.plot(x,y) 
plt.show()
```

```python
x = np.arange(0.1, 100, 0.1)
y = np.log(x)
plt.title("d)") 
plt.xlabel("x") 
plt.ylabel("f(x)") 
plt.plot(x,y) 
plt.show()
```

```python
x = np.arange(-50, 51)
y = np.exp(x)
plt.title("e)") 
plt.xlabel("x") 
plt.ylabel("f(x)") 
plt.plot(x,y) 
plt.show()
```
